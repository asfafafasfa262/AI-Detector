```python
import joblib

# 1. 載入模型與向量化工具
tfidf = joblib.load('../models/tfidf_vectorizer.pkl')
lr_model = joblib.load('../models/logistic_regression_v1.pkl')

# 2. 定義預測函數
def predict_text_traditional(text):
    # 將文字轉為 TF-IDF 向量
    text_tfidf = tfidf.transform([text])
    # 預測
    prediction = lr_model.predict(text_tfidf)[0]
    prob = lr_model.predict_proba(text_tfidf)[0]
    
    label = "AI 生成" if prediction == 1 else "人類撰寫"
    print(f"預測結果: {label} (信心度: {max(prob):.2f})")

# 測試
test_str = "This is a sample text to test the model."
predict_text_traditional(test_str)
```
